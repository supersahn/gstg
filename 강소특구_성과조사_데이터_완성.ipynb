```python
import numpy as np   # 발표자료용 4x4 하고 15, 16 제거!
import matplotlib.pyplot as plt

regions = df['지역'].unique()
indicators = ['연구소기업(건)', '기술이전(건) ', '신규창업(건)', '투자(억)', '매출액(억)', '고용(명)']

# 서브플롯 생성 (4행 4열)
fig, axes = plt.subplots(4, 4, figsize=(15, 15)) # 15, 16 제거!
axes = axes.ravel()  # 2차원 배열을 1차원으로 변환

# 각 지역별 그래프 생성
for idx, region in enumerate(regions):
    # 해당 지역 데이터 필터링
    region_data = df[df['지역'] == region]

    # 막대 그래프
    ax = axes[idx]
    x = np.arange(len(indicators))
    width = 0.35

    # 2022년, 2023년 데이터 추출
    data_2022 = [region_data[region_data['연도'] == 2022][ind].values[0] for ind in indicators]
    data_2023 = [region_data[region_data['연도'] == 2023][ind].values[0] for ind in indicators]

    # 막대 그래프 생성
    ax.bar(x - width / 2, data_2022, width, label='2022', color='skyblue')
    ax.bar(x + width / 2, data_2023, width, label='2023', color='lightcoral')

    # 그래프 꾸미기
    ax.set_title(f'{region} 강소특구 지표별 성과', pad=10, fontsize=10)
    ax.set_xticks(x)
    ax.set_xticklabels([ind.strip().split('(')[0] for ind in indicators], rotation=45)
    ax.legend()
    ax.grid(True, alpha=0.3)

# 빈 서브플롯 제거
if len(regions) < len(axes):
    for i in range(len(regions), len(axes)):
        fig.delaxes(axes[i])

plt.tight_layout()
plt.show()
```

### 2. 방사형 그래프
- 14개 특구를 한번에 비교할 수 있는 방사형 그래프로 시각화 확인

```python
# 데이터 준비 - 지역과 지표 목록 설정
regions = df['지역'].unique()
indicators = ['연구소기업(건)', '기술이전(건) ', '신규창업(건)', '투자(억)', '매출액(억)', '고용(명)']

# 2022년 데이터만
df_2022 = df[df['연도'] == 2022]

# 각 지표별 최대값으로 정규화하여 0~1 사이 값으로 변환
df_normalized = pd.DataFrame()
for indicator in indicators:
    df_normalized[indicator] = df_2022[indicator] / df_2022[indicator].max()

# 방사형 그래프 기본설정식
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='polar')

# 방사형 그래프의 각도 계산 (지표 개수만큼 균등하게 분할)
angles = np.linspace(0, 2*np.pi, len(indicators), endpoint=False)
angles = np.concatenate((angles, [angles[0]]))  # 그래프를 닫기 위해 첫 지표를 마지막에 추가 해야됨

# 각 지역별 색상(무지개 색상 사용)
colors = plt.cm.rainbow(np.linspace(0, 1, len(regions)))

# 각 지역별로 데이터를 그래프에 표시
for idx, region in enumerate(regions):
    # 해당 지역의 정규화된 값 추출
    values = df_normalized.loc[df_2022['지역'] == region, indicators].values.flatten()
    values = np.concatenate((values, [values[0]]))  # 그래프를 닫기 위해 첫 값을 마지막에 추가 해야됨

    # 선과 점으로 데이터 표시
    ax.plot(angles, values, 'o-', linewidth=2, label=region, color=colors[idx])
    # 영역 채우기 (투명도 0.1)
    ax.fill(angles, values, alpha=0.1, color=colors[idx])

# 그래프 디자인 설정
ax.set_xticks(angles[:-1])
ax.set_xticklabels([ind.strip().split('(')[0] for ind in indicators])
ax.set_title('2022년 특구별 성과 비교', pad=20, fontsize=15)
ax.grid(True)

# 범례 위치 설정 (그래프 오른쪽)
plt.legend(bbox_to_anchor=(1.3, 1.0), loc='upper right')

# 레이아웃 조정 및 그래프 표시
plt.tight_layout()
plt.show()
```

# 2023 종합 성과 지수 개발

 ## - 종합 성과 지수 계산

```python
df_2022.columns
```

```python
# 지표를 정규화, 가중 평균을 계산
def calculate_performance_index(df):
    # 정규화할 지표의 정의 (연구소기업, 기술이전, 신규창업, 투자, 매출액, 고용 데이터)
    features = ['연구소기업(건)', '기술이전(건) ', '신규창업(건)',
                '투자(억)', '매출액(억)', '고용(명)']
    df_norm = df.copy()

    # 각 지표별로 Min-Max 정규화, 정규화 공식: (x - min) / (max - min)
    # 결과값은 0에서 1 사이의 값으로 정규화, 최소값이면 0, 최대값이면 1
    for feature in features:
        min_val = df[feature].min()  # 각 지표의 최소값
        max_val = df[feature].max()  # 각 지표의 최대값
        df_norm[feature] = (df[feature] - min_val) / (max_val - min_val)  # 정규화 적용

    # 모든 지표에 대해 가중치는 동일하게
    # 정규화된 6개 지표의 평균값을 종합성과지수로 사용
    df_norm['종합성과지수'] = df_norm[features].mean(axis=1)

    return df_norm

# 2022년도 데이터만을 대상으로 종합 성과 지수 산출
df_2022_scored = calculate_performance_index(df_2022)
```

```python
# 지표를 정규화, 가중 평균을 계산
def calculate_performance_index(df):
    # 정규화할 지표의 정의 (연구소기업, 기술이전, 신규창업, 투자, 매출액, 고용 데이터)
    features = ['연구소기업(건)', '기술이전(건) ', '신규창업(건)',
                '투자(억)', '매출액(억)', '고용(명)']
    df_norm = df.copy()

    # 각 지표별로 Min-Max 정규화, 정규화 공식: (x - min) / (max - min) / 0에서 1사이 값으로 정규화, 최소값은 0-최대값은 1
    for feature in features:
        min_val = df[feature].min()  # 각 지표의 최소값
        max_val = df[feature].max()  # 각 지표의 최대값
        df_norm[feature] = (df[feature] - min_val) / (max_val - min_val)  # 정규화 적용

    # 모든 지표에 대해 가중치는 동일하게
    # 정규화된 6개 지표의 평균값을 종합성과지수로 사용
    df_norm['종합성과지수'] = df_norm[features].mean(axis=1)

    return df_norm

# 2023년도 데이터만을 대상으로 종합 성과 지수 산출
df_2023_scored = calculate_performance_index(df_2023)
```
